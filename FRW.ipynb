$$\Delta E = \dfrac{1}{2} m V^2 - \dfrac{GMm}{R} \tag{1}$$ bu denklemde sol tarafı sabit kabul edelim çünkü enerji korunur.

$$mV^2 -\dfrac{2GMm}{R} = sabit$$ 

burada yukarıdaki denklemi $m$'e bölelim 

$$V^2 - \dfrac{2GM}{R} = sabit$$

$R$ ifadesi yerine $R = a \times \Delta x$ (ölçek faktörü),  $V$ yerine ise $V = \dot{R} = \dot{a} \times \Delta x$ ($\dot{R}$ ve $\dot{a}$ zamana göre türev) yazalım

$$\dot{a}^2 \Delta x^2 - \dfrac{2GM}{a \Delta x} = sabit$$

yukarıdaki denklemde $M$ kütle ifadesini $\rho = \dfrac{M}{V}$ , $yoğunluk = \dfrac{kütle}{hacim}$ den yararlanarak $M$ yerine yoğnluk ifadesi olan $\rho$ ile yazalım yani $M = \rho \times V$ ve $V (hacim) = \dfrac{4}{3} \pi R^3$  yazalım

$$\dot{a}^2 \Delta x^2 - \dfrac{2 G \dfrac{4}{3} \pi R^3 \rho}{a \Delta x} = sabit$$
$$ \dot{a}^2 \Delta x^2 - \dfrac{8}{3}\dfrac{G \pi R^3 \rho}{a \Delta x} = sabit$$
$R$ ifadesi yerine $R = a \times \Delta x$ (ölçek faktörü), 

$$ \dot{a}^2 \Delta x^2 - \dfrac{8}{3}\dfrac{G \pi (a^3 \Delta x^3) \rho}{a \Delta x} = sabit$$
$$\dot{a}^2 \Delta x^2 - \dfrac{8}{3}{G \pi (a^2 \Delta x^2) \rho} = sabit$$
şimdi sadeliştirme yapmak için her iki tarafı $\Delta x^2$ bölelim

$$ \dot{a}^2 - \dfrac{8}{3}{G \pi (a^2) \rho} = sabit$$
şimdi ölçek faktörü ve türevini ($a, \dot{a}$) bir araya toplayalım yani her tarafı $\dfrac{1}{a^2}$ ye bölelim ve  denklemde $sabit = -k$ diyelim **$k$ ifadesi evrenin şekli ile ilgili bir ifadedir**. 

$$\left(\dfrac{\dot{a}}{a}\right)^2 - \dfrac{8}{3}{G \pi \rho} = \dfrac{-k}{a^2} \tag{FRW}$$




# Madde Baskın Model

$FRW$ denkleminde $k=0$  yazalım $\rho$ youğunluk ifadesinin önemini anlamaya çalışıyoruz

$$\left(\dfrac{\dot{a}}{a}\right)^2 - \dfrac{8}{3}{G \pi \rho} = 0$$
$\rho = \dfrac{M}{R^3} = \dfrac{M}{a^3}$ ifadesini yerine yazalım

$$\left(\dfrac{\dot{a}}{a}\right)^2 - \dfrac{8\pi G}{3}{}\dfrac{M}{a^3} = 0$$

## Türev Yaklaşımı

Burada $\dot{a}$, $a$'nın zamana göre türevi ($\frac{da}{dt}$) anlamına gelir. Denklemi yeniden düzenleyelim:

$$\left(\dfrac{\dot{a}}{a}\right)^2 = \dfrac{8\pi G}{3}{}\dfrac{M}{a^3}$$

Her iki tarafın karekökünü alalım:

$$\dfrac{\dot{a}}{a} = \sqrt{\dfrac{8\pi GM}{3a^3}}$$

Bu ifadeyi şu şekilde de yazabiliriz:

$$\dfrac{da}{dt} = \sqrt{\dfrac{8\pi GM}{3}} \cdot \dfrac{1}{a^{1/2}}$$

Değişkenleri ayırarak integrali alabiliriz:

$$a^{1/2} da = \sqrt{\dfrac{8\pi GM}{3}} dt$$

Şimdi her iki tarafın integralini alalım:

$$\int a^{1/2} da = \int \sqrt{\dfrac{8\pi GM}{3}} dt$$

Sol tarafın integrali $\frac{2}{3}a^{3/2}$ ve sağ tarafın integrali $\sqrt{\frac{8\pi GM}{3}}t + C$ olur (burada C bir integral sabiti):

$$\frac{2}{3}a^{3/2} = \sqrt{\dfrac{8\pi GM}{3}}t + C$$

$a(t)$'yi bulmak için denklemi düzenleyelim:

$$a^{3/2} = \frac{3}{2}\sqrt{\dfrac{8\pi GM}{3}}t + \frac{3}{2}C$$

$$a(t) = \left(\frac{3}{2}\sqrt{\dfrac{8\pi GM}{3}}t + \frac{3}{2}C\right)^{2/3}$$

Başlangıç koşullarını göz önüne aldığımızda, genellikle integral sabiti ihmal edilebilir. Bu durumda:

$$a(t) \propto \left(t\right)^{2/3}$$

Bu, $a(t)$'nin $t^{2/3}$ ile orantılı olduğunu gösterir.

```python
import numpy as np
import matplotlib.pyplot as plt

# Zaman dizisi
t_normal = np.linspace(0.1, 10, 400)  # Normal zaman dizisi
t_log = np.logspace(-2, 1, 400)       # Logaritmik zaman dizisi

# Ölçek faktörleri (normal zaman dizisi)
a_madde_normal = t_normal ** (2/3)
a_madde_log = t_log ** (2/3)

# Grafik oluşturma (normal değerler)
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(t_normal, a_madde_normal, label='Madde Baskın Evren: $a(t)\\propto t^{2/3}$')
plt.xlabel('Zaman (t)')
plt.ylabel('Ölçek Faktörü (a(t))')
plt.xlim((0,4))
plt.ylim((0,2))
plt.title('Normal Değerler a - t Grafiği')
plt.legend()
plt.grid(True)

# Grafik oluşturma (logaritmik değerler)
plt.subplot(1, 2, 2)
plt.plot(t_log, a_madde_log, label='Madde Baskın Evren: $a(t)\\propto t^{2/3}$')
plt.xlabel('Zaman (logaritmik) (t)')
plt.ylabel('Ölçek Faktörü (a(t))')
plt.title('Logaritmik Değerler a - t Grafiği')
plt.xscale('log')  # X eksenini logaritmik ölçeklendirme
plt.yscale('log')  # Y eksenini logaritmik ölçeklendirme
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

## Grafik yaklaşım

Amacımız,
$$
\frac{da}{dt} = \sqrt{\frac{8\pi GM}{3}} \cdot \frac{1}{\sqrt{a}}
$$
denklemini çözelim.

1. Denklemi Basitleştirme

$\sqrt{\frac{8\pi GM}{3}}$ terimi, $G$, $M$ gibi sabitleri içerdiği için, bu terimi basitleştirmek adına bir $K$ sabiti ile değiştirebiliriz:
$$
\frac{da}{dt} = \frac{K}{\sqrt{a}}
$$
Burada $K = \sqrt{\frac{8\pi GM}{3}}$.

2. Çözümün Formunu Varsayma

$a(t)$ için bir çözüm formu varsayalım. En genel formlardan biri üstel bir formdur:
$$
a(t) = Ct^n
$$
Burada:
+ $C$ bir sabittir (ölçek faktörü).
+  $n$, bulmaya çalıştığımız üsdür.

3. Türev Alma

Varsaydığımız $a(t)$ fonksiyonunun zamana göre türevini alalım:
$$
\frac{da}{dt} = nCt^{n-1}
$$

4. Orijinal Denklemle Karşılaştırma

Şimdi bu türevi orijinal denklemle karşılaştıralım:
$$
nCt^{n-1} = \frac{K}{\sqrt{Ct^n}} = KC^{-1/2}t^{-n/2}
$$

5. Üsleri Eşitleme

Bu eşitliğin sağlanabilmesi için, $t$'nin üslerinin her iki tarafta da eşit olması gerekir. Yani:
$$
n - 1 = -\frac{n}{2}
$$

6. $n$ İçin Çözüm

Bu basit cebirsel denklemi $n$ için çözelim:
$$
\begin{align*}
n + \frac{n}{2} &= 1 \\
\frac{3n}{2} &= 1 \\
n &= \frac{2}{3}
\end{align*}
$$
7. Sonuca Ulaşma

$n$'yi bulduğumuza göre, varsaydığımız çözüm formülüne geri dönebiliriz:
$$
a(t) = Ct^{2/3}
$$
Bu, $a(t)$'nin $t^{2/3}$ ile orantılı olduğunu gösterir:
$$
a(t) \propto t^{2/3}
$$

Bu yöntemde, diferansiyel denklemi doğrudan çözmek yerine, çözümün olası bir formunu varsaydık ve bu formun türevini alarak orijinal denklemle karşılaştırdık. Üsleri eşitleyerek, bilinmeyen üs $n$'yi bulduk ve böylece $a(t) \propto t^{2/3}$ sonucuna ulaştık. Bu, diferansiyel denklemleri çözmek için güçlü bir tekniktir ve özellikle denklemin analitik çözümü zor olduğunda kullanışlıdır.

# Işınım (Radyasyon) Baskın Model

Friedmann-Robertson-Walker (FRW) denklemi genel haliyle şöyledir:

$$
\left(\dfrac{\dot{a}}{a}\right)^2 - \dfrac{8\pi G}{3}\rho = \dfrac{-kc^2}{a^2} \tag{FRW}
$$

Burada:
+ $\dot{a}$ : Ölçek faktörünün zamana göre türevi (genişleme hızı)
+ $a$ : Ölçek faktörü
+ $G$ : Evrensel çekim sabiti
+ $\rho$ : Evrenin toplam enerji yoğunluğu
+ $k$ : Evrenin eğrilik sabiti (-1, 0 veya +1)
+  $c$ : Işık hızı


1. Radyasyon Baskın Evrende Basitleştirme

Radyasyon baskın evrende aşağıdaki kabuller yapılır:


 + Eğrilik terimi ihmal edilir: $\frac{-kc^2}{a^2} \approx 0$. Erken evrende enerji yoğunluğu çok yüksek olduğundan, eğrilik etkileri göz ardı edilebilir.
+ Evrenin enerji yoğunluğu büyük ölçüde radyasyondan oluşur ($\rho \approx \rho_r$) ve radyasyon yoğunluğu ölçek faktörünün dördüncü kuvvetiyle ters orantılıdır: $\rho_r \propto \frac{1}{a^4}$. Bunun nedeni, evren genişledikçe hem radyasyonun enerji yoğunluğunun azalması (hacim artışı) hem de fotonların dalga boylarının uzaması (enerji kaybı)dır.

$\rho = \rho_0 = \frac{E}{V} = \dfrac{\frac{hc}{\lambda}}{\lambda^3} = \dfrac{hc}{\lambda^4}$ 

Bu basitleştirmelerle FRW denklemi şu hale gelir:

$$
\left(\frac{\dot{a}}{a}\right)^2 \propto \frac{1}{a^4}
$$

Karekök alırsak:

$$
\frac{\dot{a}}{a} \propto \frac{1}{a^2}
$$

veya

$$
\dot{a} \propto \frac{1}{a}
$$

2. Grafik Yaklaşımı (Eğim Analizi)

Orantı işaretini bir sabitle değiştirerek denklemi şu şekilde yazabiliriz:

$$
\frac{da}{dt} = \frac{K}{a}
$$

Burada $K$ bir orantı sabitidir.

$a(t)$ için bir çözüm formu varsayalım:

$$
a(t) = Ct^n
$$

Burada $C$ bir sabittir ve $n$ bulmaya çalıştığımız üsdür.

Varsaydığımız çözümün zamana göre türevini alalım:

$$
\frac{da}{dt} = nCt^{n-1}
$$

Şimdi bu türevi basitleştirilmiş FRW denklemiyle karşılaştıralım:

$$
nCt^{n-1} = \frac{K}{Ct^n} = KC^{-1}t^{-n}
$$

Bu eşitliğin sağlanabilmesi için $t$'nin üslerinin eşit olması gerekir:

$$
n - 1 = -n
$$

Bu denklemi $n$ için çözelim:

$$
2n = 1 \implies n = \frac{1}{2}
$$

Dolayısıyla, çözüm şu şekildedir:

$$
a(t) = Ct^{1/2}
$$

Bu da, ölçek faktörünün zamanın kareköküyle orantılı olarak arttığı anlamına gelir:

$$
a(t) \propto t^{1/2}
$$


```python
import numpy as np
import matplotlib.pyplot as plt

# Zaman dizisi
t_normal = np.linspace(0.1, 10, 400)  # Normal zaman dizisi
t_log = np.logspace(-2, 1, 400)       # Logaritmik zaman dizisi


a_radyasyon_normal = t_normal ** (1/2)
a_radyasyon_log = t_log ** (1/2)


# Grafik oluşturma (normal değerler)
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(t_normal, a_radyasyon_normal, label='Radyasyon Baskın Evren: $a(t)\\propto t^{1/2}$')
plt.xlabel('Zaman (t)')
plt.ylabel('Ölçek Faktörü (a(t))')
plt.xlim((0,4))
plt.ylim((0,2))
plt.title('Normal Değerler a - t Grafiği')
plt.legend()
plt.grid(True)

# Grafik oluşturma (logaritmik değerler)
plt.subplot(1, 2, 2)
plt.plot(t_log, a_radyasyon_log, label='Radyasyon Baskın Evren: $a(t)\\propto t^{1/2}$')
plt.xlabel('Zaman (logaritmik) (t)')
plt.ylabel('Ölçek Faktörü (a(t))')
plt.title('Logaritmik Değerler a - t Grafiği')
plt.xscale('log')  # X eksenini logaritmik ölçeklendirme
plt.yscale('log')  # Y eksenini logaritmik ölçeklendirme
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

# Boşluk (Kozmolojik Sabit) Baskın Model

Boşluk baskın evrende, evrenin enerji yoğunluğunun büyük bir kısmı boşluk enerjisinden (kozmolojik sabit, Λ) kaynaklanır. Bu durumda, enerji yoğunluğu sabittir, yani ölçek faktörü *a*'ya bağlı değildir:

$$\rho = \rho_\Lambda = \text{sabit}$$

#### Doğru Yaklaşım

FRW denklemi şu şekildedir:

$$\left(\frac{\dot{a}}{a}\right)^2 - \frac{8\pi G}{3}\rho = \frac{-kc^2}{a^2}$$

Boşluk baskın evrende, enerji yoğunluğu sabit olduğundan $\rho = \rho_0 = \rho_{\Lambda}$ ve genellikle düz bir evren ($k=0$) kabul ettiğimizden, FRW denklemi şu şekilde basitleşir:

$$\left(\frac{\dot{a}}{a}\right)^2 = \frac{8\pi G}{3}\rho_\Lambda$$

Burada $\rho_\Lambda$ boşluk enerjisi yoğunluğudur. Kozmolojik sabiti Λ ile ilişkilendirmek için, şu ilişkiyi kullanırız:

$$\rho_\Lambda = \frac{\Lambda c^2}{8\pi G}$$

Bunu FRW denklemine yerine koyarsak:

$$\left(\frac{\dot{a}}{a}\right)^2 = \frac{\Lambda c^2}{3}$$

Karekök alırsak:

$$\frac{\dot{a}}{a} = H = \sqrt{\frac{\Lambda c^2}{3}} = \text{sabit}$$

Burada $H$ Hubble sabitidir (bu durumda sabit bir değerdir).

Bu denklemi çözmek için değişkenleri ayırırız:

$$\frac{da}{a} = H dt$$

Her iki tarafın integralini alırsak:

$$\int \frac{da}{a} = \int H dt$$

$$\ln(a) = Ht + \text{sabit}$$

Üs alırsak:

$$a(t) = e^{Ht + \text{sabit}} = e^{Ht}e^{\text{sabit}}$$

$e^{\text{sabit}}$'i yeni bir sabit $a_0$ ile gösterirsek:

$$a(t) = a_0e^{Ht}$$

Genellikle $a_0 = 1$ olarak alınır ve bu durumda:

$$a(t) = e^{Ht}$$

veya $H=1$ olarak seçilirse

$$a(t) \propto e^t$$

elde ederiz.



```python
import numpy as np
import matplotlib.pyplot as plt

# Hubble sabiti (örnek olarak 0.7 alalım, birimler: 1/zaman)
H = 1

# Zaman dizisi
t_normal = np.linspace(0.1, 10, 400)  # Normal zaman dizisi
t_log = np.logspace(-2, 1, 400)       # Logaritmik zaman dizisi

# Ölçek faktörleri
a_bosluk_normal = np.exp(H * t_normal)
a_bosluk_log = np.exp(H * t_log)

# Grafik oluşturma (normal değerler)
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(t_normal, a_bosluk_normal, label='Boşluk Baskın Evren: $a(t)\\propto e^{Ht}$')
plt.xlabel('Zaman (t)')
plt.ylabel('Ölçek Faktörü (a(t))')
plt.xlim((0,4))
plt.ylim((0,2))
plt.title('Normal Değerler a - t Grafiği')
plt.legend()
plt.grid(True)

# Grafik oluşturma (logaritmik değerler)
plt.subplot(1, 2, 2)
plt.plot(t_log, a_bosluk_log, label='Boşluk Baskın Evren: $a(t)\\propto e^{Ht}$')
plt.xlabel('Zaman (logaritmik) (t)')
plt.ylabel('Ölçek Faktörü (a(t))')
plt.title('Logaritmik Değerler a - t Grafiği')
plt.xscale('log')  # X eksenini logaritmik ölçeklendirme
plt.yscale('log')  # Y eksenini logaritmik ölçeklendirme
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```


#### Sonuç

Boşluk baskın evrende, enerji yoğunluğu sabit olduğundan, FRW denklemi basit bir diferansiyel denkleme dönüşür ve çözümü $a(t) \propto e^t$ şeklinde üstel bir genişleme verir. Bu, evrenin hızlanarak genişlediği anlamına gelir.


# Evrenin Farklı Dönemleri ve Genişleme Davranışları (Özet):

Evrenin genişlemesi, Friedmann-Robertson-Walker (FRW) denklemleri ile açıklanır. Bu denklemlerin çözümleri, evrenin baskın enerji türüne göre farklılık gösterir. Başlıca üç dönem vardır:

1. **Radyasyon Baskın Evren:**
    
    - **Dönem:** Evrenin ilk zamanları, Büyük Patlama'dan hemen sonraki dönem.
    - **Baskın Enerji:** Radyasyon (fotonlar, relativistik parçacıklar).
    - **Genişleme Davranışı:** a(t)∝t1/2 (Ölçek faktörü zamanın kareköküyle orantılı).
    - **Genişleme Hızı:** Nispeten hızlı, ancak yavaşlayan bir genişleme.
2. **Madde Baskın Evren:**
    
    - **Dönem:** Radyasyon döneminden sonra, madde yoğunluğunun radyasyonu aştığı dönem.
    - **Baskın Enerji:** Madde (baryonlar, karanlık madde).
    - **Genişleme Davranışı:** a(t)∝t2/3 (Ölçek faktörü zamanın 2/3 kuvvetiyle orantılı).
    - **Genişleme Hızı:** Radyasyon baskın döneme göre daha yavaş, ancak hala yavaşlayan bir genişleme.
3. **Boşluk (Karanlık Enerji) Baskın Evren:**
    
    - **Dönem:** Günümüz evreni ve gelecekteki evren.
    - **Baskın Enerji:** Boşluk enerjisi (kozmolojik sabit).
    - **Genişleme Davranışı:** a(t)∝eHt (Ölçek faktörü üstel olarak artar).
    - **Genişleme Hızı:** Hızlanan bir genişleme.

**Genel Değerlendirme:**

Evrenin genişlemesi, farklı dönemlerde farklı dinamikler göstermiştir. Başlangıçta radyasyonun baskın olduğu dönemde hızlı, ancak yavaşlayan bir genişleme vardı. Daha sonra madde baskın döneme geçildiğinde genişleme hızı yavaşlamaya devam etti. Ancak, günümüzde karanlık enerjinin baskın hale gelmesiyle birlikte evrenin genişlemesi _hızlanmaktadır_.

Bu hızlanan genişleme, kozmolojik sabitin veya daha genel olarak karanlık enerjinin varlığına güçlü bir kanıttır. Karanlık enerji, evrenin enerji içeriğinin yaklaşık %68'ini oluşturur ve evrenin gelecekteki evrimi üzerinde belirleyici bir role sahiptir.

**Özet Tablo:**

| Dönem                  | Baskın Enerji   | Genişleme Davranışı   | Genişleme Hızı    |
| ---------------------- | --------------- | --------------------- | ----------------- |
| Radyasyon Baskın Evren | Radyasyon       | $a(t)\propto t^{1/2}$ | Hızlı, Yavaşlayan |
| Madde Baskın Evren     | Madde           | $a(t)\propto t^{2/3}$ | Yavaş, Yavaşlayan |
| Boşluk Baskın Evren    | Boşluk Enerjisi | $a(t)\propto e^{Ht}$  | Hızlanan          |

Bu tablo, evrenin farklı dönemlerindeki genişleme davranışlarını ve baskın enerji türlerini net bir şekilde özetlemektedir. Bu bilgiler, evrenin geçmişini, bugününü ve geleceğini anlamak için temel oluşturmaktadır.

```python
import numpy as np
import matplotlib.pyplot as plt

# Hubble sabiti (örnek olarak 0.7 alalım, birimler: 1/zaman)
H = 1

# Zaman dizisi
t_normal = np.linspace(0.1, 10, 400)  # Normal zaman dizisi
t_log = np.logspace(-2, 1, 400)       # Logaritmik zaman dizisi

# Ölçek faktörleri (normal zaman dizisi)
a_radyasyon_normal = t_normal ** (1/2)
a_madde_normal = t_normal ** (2/3)
a_bosluk_normal = np.exp(H * t_normal)

# Ölçek faktörleri (logaritmik zaman dizisi)
a_radyasyon_log = t_log ** (1/2)
a_madde_log = t_log ** (2/3)
a_bosluk_log = np.exp(H * t_log)

# Grafik oluşturma (normal değerler)
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(t_normal, a_radyasyon_normal, label='Radyasyon Baskın Evren: $a(t)\\propto t^{1/2}$')
plt.plot(t_normal, a_madde_normal, label='Madde Baskın Evren: $a(t)\\propto t^{2/3}$')
plt.plot(t_normal, a_bosluk_normal, label='Boşluk Baskın Evren: $a(t)\\propto e^{Ht}$')
plt.xlabel('Zaman (t)')
plt.ylabel('Ölçek Faktörü (a(t))')
plt.xlim((0,4))
plt.ylim((0,2))
plt.title('Normal Değerler a - t Grafiği')
plt.legend()
plt.grid(True)

# Grafik oluşturma (logaritmik değerler)
plt.subplot(1, 2, 2)
plt.plot(t_log, a_radyasyon_log, label='Radyasyon Baskın Evren: $a(t)\\propto t^{1/2}$')
plt.plot(t_log, a_madde_log, label='Madde Baskın Evren: $a(t)\\propto t^{2/3}$')
plt.plot(t_log, a_bosluk_log, label='Boşluk Baskın Evren: $a(t)\\propto e^{Ht}$')
plt.xlabel('Zaman (logaritmik) (t)')
plt.ylabel('Ölçek Faktörü (a(t))')
plt.title('Logaritmik Değerler a - t Grafiği')
plt.xscale('log')  # X eksenini logaritmik ölçeklendirme
plt.yscale('log')  # Y eksenini logaritmik ölçeklendirme
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```


| Evren Modeli         | Genişleme Davranışı ($a(t)$)   | Yoğunluk Davranışı ($\rho(t)$)    |
|----------------------|-------------------------------|----------------------------------|
| Işınım Baskın Evren  | $a(t) \propto t^{1/2}$        | $\rho \propto a^{-4}$            |
| Madde Baskın Evren   | $a(t) \propto t^{2/3}$        | $\rho \propto a^{-3}$            |
| Boşluk Baskın Evren  | $a(t) \propto e^{Ht}$         | $\rho = \text{constant}$         |



1. **Işınım Baskın Evren**:
   - **Genişleme Davranışı**: $a(t) \propto t^{1/2}$ - Ölçek faktörü zamanın karekökü ile orantılı olarak artar.
   - **Yoğunluk Davranışı**: $\rho \propto a^{-4}$ - Yoğunluk ölçek faktörünün dördüncü kuvvetiyle ters orantılı olarak azalır.

2. **Madde Baskın Evren**:
   - **Genişleme Davranışı**: $a(t) \propto t^{2/3}$ - Ölçek faktörü zamanın 2/3 kuvveti ile orantılı olarak artar.
   - **Yoğunluk Davranışı**: $\rho \propto a^{-3}$ - Yoğunluk ölçek faktörünün küpü ile ters orantılı olarak azalır.

3. **Boşluk Baskın Evren**:
   - **Genişleme Davranışı**: $a(t) \propto e^{Ht}$ - Ölçek faktörü Hubble sabiti \(H\) ile üssel olarak artar.
   - **Yoğunluk Davranışı**: $\rho = \text{constant}$ - Yoğunluk sabit kalır (karanlık enerji yoğunluğu zamanla değişmez).
